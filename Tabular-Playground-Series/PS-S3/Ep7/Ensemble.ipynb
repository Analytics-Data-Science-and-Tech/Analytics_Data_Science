```python
import pandas as pd 
import numpy as np

lgb = pd.read_csv('LightGBM_Leakage_3_best.csv')
lgb.columns = ['id', 'pred1']

xgb = pd.read_csv('XGBoost_Leakage_2_42_best.csv')
xgb.columns = ['id', 'pred2']

cat = pd.read_csv('CatBoost_Leakage_2_42_.csv')
cat.columns = ['id', 'pred3']

hist = pd.read_csv('Hist_Leakage_3_20_features.csv')
hist.columns = ['id', 'pred4']

ensemble = pd.merge(lgb, xgb, on = 'id')
ensemble = pd.merge(ensemble, cat, on = 'id')
ensemble = pd.merge(ensemble, hist, on = 'id')
ensemble.head()
```

```python
ensemble['booking_status'] = (ensemble['pred1'] + ensemble['pred2'] + ensemble['pred3'] + ensemble['pred4']) / 4
ensemble.drop(columns = ['pred1', 'pred2', 'pred3', 'pred4'], axis = 1, inplace = True)
ensemble.head()
```

```python
ensemble.to_csv('ensemble_3.csv', index = False)
```

# Weighted Average

```python
wtot = 0.92386 + 0.92368 + 0.92365
w1 = 0.92386 / wtot
w2 = 0.92368 / wtot
w3 = 0.92365 / wtot

lgb = pd.read_csv('LightGBM_Leakage_3_best.csv')
lgb.columns = ['id', 'pred1']

xgb = pd.read_csv('XGBoost_Leakage_2_42_best.csv')
xgb.columns = ['id', 'pred2']

cat = pd.read_csv('CatBoost_Leakage_2_42_.csv')
cat.columns = ['id', 'pred3']

ensemble = pd.merge(lgb, xgb, on = 'id')
ensemble = pd.merge(ensemble, cat, on = 'id')
ensemble.head()
```

```python
ensemble['booking_status'] = w1*ensemble['pred1'] + w2*ensemble['pred2'] + w3*ensemble['pred3']
ensemble.drop(columns = ['pred1', 'pred2', 'pred3'], axis = 1, inplace = True)
ensemble.head()
```

```python
ensemble.to_csv('ensemble_1.csv', index = False)
```
