```python
import pandas as pd 
import numpy as np

## Reading data-files
kaggle = pd.read_csv('kaggle_submission.csv')
kaggle.columns = ['Id', 'kaggle']

oscar_1 = pd.read_csv('XGB_Reg_FE_2.csv')
oscar_1.columns = ['Id', 'oscar_1']

oscar_2 = pd.read_csv('XGB_Reg_FE_4.csv')
oscar_2.columns = ['Id', 'oscar_2']

preds = pd.merge(kaggle, oscar_1, on = 'Id')
preds = pd.merge(preds, oscar_2, on = 'Id')
preds.head()
```

```python
preds['Numb_of_Uniques'] = preds.drop(columns = 'Id', axis = 0).nunique(axis = 1)
preds['quality'] = preds[['kaggle', 'oscar_1', 'oscar_2']].mode(axis = 1)
preds.head(10)
```

```python
preds[preds['Numb_of_Uniques'] > 1]
```

```python
preds['quality'].value_counts()
```

```python
preds = preds[['Id', 'quality']]
preds.head()
```

```python
preds.to_csv('ensemble_comparisson.csv', index = False)
```
